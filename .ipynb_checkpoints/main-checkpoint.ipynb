```python
import sys
from PyQt6.QtWidgets import QApplication, QMainWindow, QPushButton, QVBoxLayout, QWidget, QLabel, QFileDialog
from PyQt6.QtCore import Qt

class WhisperApp(QMainWindow):
    """
    Main application window for the Lyrics Transcriber.
    Inherits from QMainWindow to provide a standard application framework.
    """
    
    def __init__(self):
        """Initializes the window, UI components, and layout."""
        super().__init__()

        # --- Window Configuration ---
        self.setWindowTitle("AI Lyrics Transcriber (PyQt6)")
        # Set window size: setGeometry(x, y, width, height)
        self.setGeometry(100, 100, 600, 400)

        # --- Layout Architecture ---
        # QMainWindow requires a central widget to hold other UI elements
        self.central_widget = QWidget()
        self.setCentralWidget(self.central_widget)
        
        # Use QVBoxLayout to stack widgets vertically
        self.layout = QVBoxLayout()
        self.central_widget.setLayout(self.layout)

        # --- UI Element Creation ---
        # Main title label with CSS-style formatting
        self.title_label = QLabel("Whisper Lyrics Extractor")
        self.title_label.setAlignment(Qt.AlignmentFlag.AlignCenter)
        self.title_label.setStyleSheet("font-size: 20px; font-weight: bold; margin: 20px;")
        
        # Label to provide feedback on selected files (starts empty)
        self.file_label = QLabel("No file selected")
        self.file_label.setWordWrap(True) # Ensures long file paths don't break the layout
        self.file_label.setAlignment(Qt.AlignmentFlag.AlignCenter)

        # Main action button to trigger the file browser
        self.select_btn = QPushButton("Select MP3 File")
        self.select_btn.setFixedSize(200, 40)
        
        # Connect the button click signal to the open_file_dialog function (Signal/Slot mechanism)
        self.select_btn.clicked.connect(self.open_file_dialog)

        # --- Adding Widgets to the Screen ---
        self.layout.addWidget(self.title_label)
        self.layout.addWidget(self.select_btn, alignment=Qt.AlignmentFlag.AlignCenter)
        self.layout.addWidget(self.file_label)
        
        # addStretch() pushes all content to the top for a cleaner look
        self.layout.addStretch()

    def open_file_dialog(self):
        """
        Triggers a native Windows/OS file explorer window.
        Allows the user to select an audio file and updates the UI with the path.
        """
        # getOpenFileName returns a tuple: (path_to_file, filter_used)
        file_path, _ = QFileDialog.getOpenFileName(
            self, 
            "Open Audio File", 
            "", # Start directory (empty means current)
            "Audio Files (*.mp3 *.wav *.m4a);;All Files (*)" # Filters
        )
        
        # Only update the label if the user actually picked a file (didn't click Cancel)
        if file_path:
            self.file_label.setText(f"Selected: {file_path}")
            # Log the selection to the console for debugging
            print(f"DEBUG: User selected file -> {file_path}")

# --- Application Entry Point ---
if __name__ == "__main__":
    # Standard boilerplate for starting a PyQt6 application
    app = QApplication(sys.argv)
    
    # Instantiate the class we built above
    window = WhisperApp()
    
    # Render the window on screen
    window.show()
    
    # Start the event loop (keeps the app running until closed)
    sys.exit(app.exec())
```
