```python
import whisper
import os

# Load the model (Options are: 'tiny', 'base', 'small', 'medium', 'large')
print("Loading Whisper model... (this may take a moment the first time)")
model = whisper.load_model("base")

# Path to the file 
# 'r' is used before the string to handle Windows backslashes properly.
audio_path = r"C:\Users\byaku\Desktop\ki-song-pop-vocals-448727.mp3"

if os.path.exists(audio_path):
    print(f"Transcribing: {audio_path}")
    
    # Run the transcription
    # set verbose=True so you can see the lyrics appear in real-time.
    result = model.transcribe(audio_path, verbose=True)
    
    # Final Output
    print("\n--- DONE ---")
    print("Full Lyrics Output:")
    print(result["text"])
else:
    print(f"Error: The file at {audio_path} was not found.")
```
